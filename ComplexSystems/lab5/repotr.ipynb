**5.1** $\ddot{x} = x - x^2$.

а) Найдите и классифицируйте точки равновесия

Приведем диффиренциальное уравнение второго порядка к системе первого порядка
$$
        \begin{cases}
            \dot{x} = y \\
            \dot{y} = x - x^2
        \end{cases}
$$
Особые точки системы $(0, 0)$ и $(0, 1)$.

$$
   J =  \begin{bmatrix}
        0 & 1 \\
        1 - 2x & 0        
    \end{bmatrix} \Rightarrow p = 0, q = 2x - 1
$$

$$ 
   J(0, 0):  p = 0, q = - 1 \Rightarrow седло \\
   J(0, 1):  p = 0, q = 1 \Rightarrow центр
$$


б) Нарисуем фазовый портрет



в) Для нахождения границы области замкнутых траекторий требуется найти уравнение, описывающее закон сохранения энергии в системе. Для этого проинтегрируем уравнение методом разделения переменных(поделим одно уравнение системы на другое, соберем переменные по разные стороны знака равенства и проинтегрируем. После, вынесем неизвестную константу в одну часть).
$$
    \begin{array}{c}
        (x - x^2)dx = y dy \\
        \frac{x^2}{2} - \frac{x^3}{3} - \frac{y^2}{2} = C
    \end{array}
$$

Так как в уравнении присутствует неизвестный параметр, вернемся к предыдущему графику и обратим внимание на фазовые траектории. Очевидно, что необходимая нам граница должна касаться точки $(0,0)$ - центра седла, так как взяв орбиту чуть больше получим область, из которой точки будут двигаться на бесконечность, а взяв чуть меньше, оставим область в правом секторе седла, которая нам визуально подходит. Подставив эту точку в уравнение получим $C = 0$.
