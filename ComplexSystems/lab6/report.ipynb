9.1.10 φ имеет устойчивый цикл {1 􏰩→ 5 􏰩→ 2 􏰩→ 4 􏰩→ 3 􏰩→ 1}.

По данным точкам построим понятно, через какие точки графика должна искомая функция ({1, 5}, {5, 2}, {2, 4}, {4, 3}, {3, 1}). Интерполируем ее полиномом (с помощью Mathematica) и построим график:

Фактически, данный график вычеслен численно, следовательно, можно считать его численным докозательством). Для теоретической проверки применим функцию к своему результату начиная с единицы 5 раз и получим исходную последовательность:


9.5.4  $x_{n+1} = r*tg(x_n)$ (Мешанина)

Функция нечетна относительно r, поэтому построим только положительную часть графика. При r < 1 не имеем циклов (одна устойчивая точка)

Без ограничения вычислений получим большие значения (вплоть до 1^7) точек цикла. Глядя на график и эти значения кажется, что они не ограничены ничем. Поэтому, ограничим некоторую область r и x и получим следующий график:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(r, x):
    return r * np.tan(x)

n = 10000
xm, ym = 0., 100.
r = np.linspace(xm, ym, n)
iterations = 1000
last = 100
x = 1e-5 * np.ones(n)
fig, (ax1) = plt.subplots(1, 1, figsize=(8, 9), sharex=True)

for i in range(iterations):
    x = f(r, x)
    if i >= (iterations - last):
        ax1.plot(r, x, ',k', alpha=.25)
        
ax1.set_xlim(xm, ym)
ax1.set_ylim(-100, 100)
plt.show()
```

Видно, что иксы некоторым (потенциально, Гауссовским) распределением размазаны по бесконечной плоскости для $x \in [1, +\infty]$ с наибольшей плотностью в окрестности точки (1, 0).   
Отдельно заметим, что изменение начальных условий не влияет на картину бифуркационной диаграммы.
