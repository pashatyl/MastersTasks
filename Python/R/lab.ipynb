```r
b1 <- abs(13-MY_NUMBER)+2
bn <- 100
name_len = 6

ratio <- (bn/b1)^(1/(name_len-1))

geomSeq <- function(start,ratio,begin,end){
  begin=begin-1
  end=end-1
  return(start*ratio**(begin:end))
}

v2 <- geomSeq(b1, ratio, 1, name_len)
v2
```

```r
v3 <- sample(c(v1, v2), 3)
v3
```

```r
alph <-""
```

```r
swapLetters <- function(s, alph){
  N = nchar(alph)
  ans = ""
  for (j in 1:nchar(s)){
      letter <- substr(s, j,j)
      n <- gregexpr(letter, alph)[[1]]
      new_letter_ind <- ((j * (n - 1)) %% N)+1
      ans <- paste(ans, substr(alph, new_letter_ind, new_letter_ind), sep='')
  }
  return(ans)
}
```

```r
swapLetters("", alph)
```
