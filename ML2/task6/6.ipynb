## Лабораторная работа 6



Имеем: массив пар $(x, y), x\in\mathbb{R}, y\in\{0, 1\}$ (отсортированный по х, иначе отсортируем не более чем за $O(mlogm)$  
Ищем: отрезок $[x_1, x_2]$, минимизирующий эмпирический риск(сумму ...)
Очевидно, что нам потребуется сделать полный перебор пар точек - $O(m^2)$, однако вычисление эмпирического риска для произвольного отрезка занимает $O(m)$, а надо бы за $O(1)$. И здесь нам помогут комулятивные суммы! (по аналогии с домашним заданием из 1 семестра)
1. Проинициализируем коммулятивные суммы - +1 за метку 1 и -1 за 0
2. Переберем все пары точек, запомнив ту, на которой достигается максимум суммы
3. В случае отрицательной суммы возьмем отрезок, не включающий элементов выборки
Отрезок с найденным максимумом и будет ответом задачи

На словах: при выполнении гипотезы о реализуемости нам достаточно проверить "крайние" точки множества, то есть мы тратим $O(n)$ операций на поиск этих самых крайних точек. В агностическом случае нам потребуется перебрать уже все возможные подмножества множества примеров (полный перебор), для что понадобится 􏰁$C_m^n \leq m^{O(n)}$ элементов множетва.  
Итого получаем: $ m^{O(n)} * O(mn) = O(nm^{O(n)})$

Построим следующую обучающую выборку S:
$$x_i = -sign(b_i) A[i]$$
$$y_i = -sign(b_i)$$  
Теперь надо показать, что $(w, b) \in \mathbb{R}^n x \mathbb{R}$ является ERM гипотезой для $S$ и для $HS_n$ тогда, когда $w$ является решением MaxFS:  
$$h_w(x_i) = y_i \Leftrightarrow y_i( \langle w, x_i \rangle + b) > 0  
\Leftrightarrow \langle w, A[i] \rangle - b_i > 0$$  
А так как ERM гипотезой можно назвать ту, которая минимизирует количество неправильных классификаций, то отображение можно считать построенным, а значит и задача с $HS_n$ является NP-полной
