 1. Проверим данное утверждение на точках $1, -1, 0$, что позволит легко подбирать необходимый знак выбирая необходимую компонету. В качестве выборки возьмем следующую(путем нехитрого подбора): $S=\{((-1, 1),-1), ((1,-1),1)\}$. Подберем вектора $w'=(-1, 0)$ и $w=(0, 1)$, которые ошибаются на всей выборке $L_S(w)=L_S(w')=1$. $\Vert w - w'\Vert\leq\epsilon=\sqrt2$, получаем локальный минимум w. Найдем такой вектор $w''$, что $L_S(w'')\leq L_S(w)$. Возьмем $w''=(0, -1)$.  
Результатом классификации будет вектор $(-1, 1)$, т.е. $L_S(w'')=0<L_S(w)$, таким образом w является локальным минимумом, однако пример показывает, что существует отличный от нее глобальный минимум.

3. Рассмотрим функцию $g(x) = max\{0, 1-x\}$. 
По определению липшицевости $|g(x) - g(y)|=max\{0, 1-x\} - max\{0, 1-y\}$. Ограничим сверху эту разность максимальным значением при рассмотрении на промежутках:     
$$|g(x) - g(y)| \leq |y - x| = \leq |x - y|$$
Из лекции почерпнем, что функция $y\langle w, x\rangle$ - $\Vert x \Vert$-липшицева, а композиция двух липшицевых функций - липшицева(по произведению). Так же ограничим $\Vert x \Vert \leq R$ по определению шара.  
Получаем, что $g(x) = max\{0, 1-\langle w, x\rangle\}$ - $R$-липшицева
