        1 & 1 & 0 & \dots & 0 \\
        1 & 0 & 1 & \dots & 0\\
        1 & \vdots & \vdots &  \ddots & 0\\
        1 & 0 & 0 & \dots  & 1
    \end{pmatrix} $$    
    Так как Х - обратимая, мы всегда можем выбрать $w = X^T y$, который приведет $ w^Tx $ в необходимый знак
    
    $VCdim(H) \leq d+2$:
    
    Очевидно, что взяв систему из d+1 уравнений и d+2 переменной мы получим линейную зависимость, а значит не сможем покрыть множество решений своим множеством гипотез

3. а) Возьмем в качестве $С$ базис пространства, где $$x_i^C=(0,...,0,1^{(i)},0,...,0), ∀i= \overline{1,n}$$
Пусть $y_C=(y_1^C,y_2^C,...,y_n^C), y_j^C=\{0,1\}, ∀j= \overline{1,n}$ - значение произвольной функции из $H_C$.
Покажем, что для получения необходимых значений на наборе $С$ найдется функция из $H$.

    Выберем множество 
    $$I_C=\{i|y_i^C=1\}$$
    Значит, $h_{I_C}(x_j^C) = 1$, когда $j \in I_C$. Следовательно, семейство $H_C$ таких функций разукрашивает $C$ и 
    $$VCdim(H)\geq n$$

    б) $|X| = n$, значит $|H| \leq 2^n$
    $$VCdim(H) \leq log_2(|H|)$$
    $$VCdim(H) \leq n$$
